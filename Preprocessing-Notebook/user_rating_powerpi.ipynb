# Preprocessing Notebook of Customer Rating Project

## Import needed liberaries and read data

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline 
```

```python
data=pd.read_csv("/content/apple_iphone_11_reviews.csv")
```

```python
data.head()
```

## preprocessing

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.shape
```

```python
data[['NumOfHelpfulCount', 'Text']] = data['helpful_count'].str.split(' ', n=1, expand=True)
```

```python
data.head()
```

```python
data.drop(['Text','helpful_count'], axis =1 ,inplace=True)
```

```python
data.head()
```

```python
data.info()
```

```python
sns.pairplot(data)
```

```python
data['reviewed_at']=pd.to_datetime(data['reviewed_at'])
```

```python
data.columns
```

```python
data.info()
```

```python
df=data.drop([ 'product','url', 'review_country',
       'reviewed_at', 'review_text', 'review_rating', 'product_company',
       'profile_name', 'review_title', 'NumOfHelpfulCount'],axis=1)
attributs = df.columns.values
names = ['index', 'total_comments']
attributs
```

```python
def showALlBoxPlot(df, msg):
    fig, axs = plt.subplots(2,2, figsize=(10,7))
    fig.suptitle(msg)
    fig.subplots_adjust(wspace=.2, hspace=.2)

    idx = 0
    for atr, ax in zip(attributs, axs.ravel()):
        sns.boxplot(data = data[atr], ax=ax)
        ax.set_title(names[idx])
        idx+=1

    

showALlBoxPlot(data, "Box plot Summary to show Outliers")
```
